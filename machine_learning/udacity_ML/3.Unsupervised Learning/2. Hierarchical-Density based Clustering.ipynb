## Complete link 

Complete link is similar to the single link clustering but the only difference is when you want to add a new point to the cluster you don't take the distance of the closest point of the cluster to the candidate point but the farthest point of the cluster.  

![c_l_1.png](pics/c_l_1.png)

![c_l_1.png](pics/c_l_2.png)

This sort of caclulation creates compact clusters. One challenge with complete link (like single list) it looks at one point.
